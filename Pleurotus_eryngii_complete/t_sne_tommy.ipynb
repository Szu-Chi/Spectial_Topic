```python
path='/home/tommy/Documents/DATA/spectral_curve_v1/B/*'
folder_path_list = sorted(glob.glob(path))
del path
data_n=len(folder_path_list)
```

```python
arr=np.ones((data_n,300))
for i in range(data_n):
    arr[i] = np.load(folder_path_list[i])
arr.shape
```

```python
label_value = 0
data_label = np.zeros(data_n, dtype=np.int8)
check = folder_path_list[0][-16:-12]
for i in range(data_n):
    if folder_path_list[i][-16:-12] != check:
        label_value += 1
        check = folder_path_list[i][-16:-12]
    data_label[i] = label_value
print(data_label)
print(len(data_label))
for i in range(label_value):
    print(sum((data_label==i)))
```
